```python
import urllib.request as req
url = "https://www.ptt.cc/bbs/movie/index.html"
#建立一個request 物件
# 因為如果單純要求網頁容易被拒絕 所以要讓自己看起來像個人類
# 網頁 >開發人員工具>網頁 >標頭>右邊格子的標頭 
request = req.Request(url, headers ={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"
})

with req.urlopen(request) as response:
  data = response.read().decode("utf-8")
```

```python
print(data)
```

#解析資料

```python
#解析資料 需要用到 beautifulsoup4 協助解析
import bs4
```

```python
root = bs4.BeautifulSoup (data, "html.parser")  #告訴她剛剛下載下來的是html
```

```python
print(root)
```

```python
print(root.title.string) #抓到標籤的文字
```

#抓文章的標題

```python
titles = root.find("div", class_="title") #搜尋 class = "title"的div 標籤
```

```python
print(titles)
```

```python
#如果只要文字
print(titles.a.string) #標題有A </a>
```

#抓所有的標題


```python
titles_all = root.find_all("div", class_="title") #搜尋 class = "title"的div 標籤
```

```python
print(titles_all)
```

```python
for title in titles_all:
  if title.a != None: #如果標題有A就把他列出來
   print (title.a.string)
```
