```python
import time
import urllib.request as request
import json
import requests
import csv
import pandas as pd
import requests
from requests.exceptions import RequestException
import urllib.error as error

for i in range(8,11):
    id1 = []
    z = str(i)
    path ="D:/Users/PSFD/Desktop/工商登記資料/company_n"+z+".csv"   #這邊有commerce 的不同編號
    with open(path, "r", newline= "", encoding="utf-8") as number:   #製作統一編號的list
        rows = csv.reader(number, delimiter=',')
        headers = next(rows)
        for row in rows:
            id1.append(row)

    all_data1 = []
    count = 0
    start_time = time.time()  #開始時間
    for k in id1 :    
        count = count + 1  #製作計時器
        if count % 1000 == 0:
            elapsed_time = time.time() - start_time  # 計算時間經過
            print(f"Processed {count} records in {elapsed_time:.2f} seconds")
            time.sleep(100)
        src = "https://data.gcis.nat.gov.tw/od/data/api/5F64D864-61CB-4D0D-8AD9-492047CC1EA6?$format=json&$filter=Business_Accounting_NO%20eq%20"+ k[0] +"&$skip=0&$top=50"  #將統一編號區隔出來
        success = catch(src)
        while not success:  # 如果请求失败，等待1小时再重新尝试
            localtime = time.localtime()  #打出現在時間
            result = time.strftime("%Y-%m-%d %I:%M:%S %p", localtime)
            print(f"Waiting a half hour before retrying {src}")
            print(result)
            time.sleep(1800)  # 等待半小时
            success = catch(src)
    df = pd.json_normalize(all_data1)
    df.to_excel(r'D:\Users\PSFD\Desktop\工商登記資料\company_應用一\apply1_'+z+'.xlsx', index=False) #輸出成excel  # excel 請使用不同編號
```
