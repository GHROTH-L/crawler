```python
#!/usr/bin/env python
# coding: utf-8

# ## 請先下載最新的webdriver
# https://googlechromelabs.github.io/chrome-for-testing/#stable
# 並將其與python檔案放在同一個資料夾中
# ## 請使用登入中研院vpn

# In[322]:


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
import time
import random



# In[323]:


def page2(i,j):
    if i == 4:   #第二區塊因為只有一個，所以他的tr沒有數字
        xpath = "/html/body/center/table[2]/tbody/tr[{}]/td/table/tbody/tr/td[2]/a".format(i)
    else:
        xpath = "/html/body/center/table[2]/tbody/tr[{}]/td/table/tbody/tr[{}]/td[2]/a".format(i,j)
    start_button = driver.find_element("xpath", xpath)
    start_button.click()
    chall_button = driver.find_element("xpath", "/html/body/center[1]/form/input[4]")
    chall_button.click()
    chall_button = driver.find_element("xpath", "/html/body/center[1]/form/input[4]")
    chall_button.click()
    takein_button = driver.find_element("xpath","/html/body/center[1]/form/input[6] ")
    takein_button.click()
    window_handles = driver.window_handles
    return  driver.switch_to.window(window_handles[1])
    time.sleep(10)


# In[324]:


def data(driver):

    #年資料
    st_y = driver.find_element("xpath", "/html/body/center/table/tbody/tr[1]/td/form/center[2]/form/select[1]")
    s_st = Select(st_y)
    s_st.select_by_index(1) #抓取第一筆

    ed_y = driver.find_element("xpath","/html/body/center/table/tbody/tr[1]/td/form/center[2]/form/select[2]")
    s_ed = Select(ed_y)
    edy_options = len(s_ed.options)
    s_ed.select_by_index(edy_options -2) #2023

    get_data = driver.find_element("xpath", "/html/body/center/table/tbody/tr[1]/td/form/center[2]/form/input[2]")
    get_data.click()  #點開後會直接下載

    window_handles = driver.window_handles
    driver.switch_to.window(window_handles[-1])  #更換到最新頁，如果最新頁有alert（沒有資料）就按確認
    random_wait_time = random.uniform(1, 5)
    time.sleep(random_wait_time)
    try:
        driver.switch_to.alert.accept()
    except:
        driver.close()


    #月資料
    driver.switch_to.window(window_handles[1])
    st_ym = driver.find_element("xpath", "/html/body/center/table/tbody/tr[3]/td/form/center[2]/form/select[1]")
    s_stm = Select(st_ym)
    s_stm.select_by_index(1)

    ed_ym = driver.find_element("xpath","/html/body/center/table/tbody/tr[3]/td/form/center[2]/form/select[3]")
    s_edm= Select(ed_ym)
    edym_options = len(s_edm.options)
    s_edm.select_by_index(edym_options -2)

    get_data = driver.find_element("xpath", "/html/body/center/table/tbody/tr[3]/td/form/center[2]/form/input[2]")
    get_data.click()  #點開後會直接下載

    window_handles = driver.window_handles
    driver.switch_to.window(window_handles[-1]) #更換到最新頁，如果最新頁有alert（沒有資料）就按確認
    random_wait_time = random.uniform(1, 5)
    time.sleep(random_wait_time)
    try:
        driver.switch_to.alert.accept()
    except:
        driver.close()
    driver.switch_to.window(window_handles[1])

    for z in range(1,4):
        dxpath = "/html/body/center/table/tbody/tr[{}]/td/form/center[2]/input[2]".format(z)
        delet_data = driver.find_element("xpath",dxpath)
        delet_data.click()
        driver.switch_to.alert.accept()
        time.sleep(5)

    driver.close()
    driver.switch_to.window(window_handles[0])
    random_wait_time = random.uniform(1, 5)
    time.sleep(random_wait_time)
    if i % 14 == 0:  # 如果 i 能被 20 整除
        driver.refresh()  # 先進行網頁重新整理  處理大概30筆就會被排除
        driver.back()  # 返回上一頁
    else:
        driver.back()





# In[325]:


if __name__ == '__main__':
    options = Options()
    options.chrome_executable_path = "D:\PYTHON"
    driver = webdriver.Chrome(options = options)
    driver.get("資料庫網址")
    section = {}
    for i in range(2,13,2):  #計算出mainpage 的區塊中有幾個可點擊連結 #如果要調整的話，請調整最前面的數字（2 4 6 8 10 分別表示不同1-5的section)
        xpath1 = "/html/body/center/table[2]/tbody/tr[{}]/td/table/tbody".format(i)
        tbody_element = driver.find_element("xpath", xpath1)
        tr_elements = tbody_element.find_elements("tag name", "tr")
        section[i] = len(tr_elements)
        for j in range(1, section[i]+1):
            page2(i,j)
            data(driver)
            random_wait_time = random.uniform(10, 30)
            time.sleep(random_wait_time)
```
