```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
# 生成ttl
def gen_url(url, teams, years):
  ttl = []
  for team in teams:
    for year in years:
      newurl = url.format(team, year)
      ttl.append([team, year, newurl])
  return ttl
```

```python
#生成選手
def players(soup,team,year):
  players = []
  table = soup.find(id="roster") #找到表格
  for row in table.find("tbody").find_all("tr"):
    no = row.th.text
    cols = row.findAll("td")
      #球隊, 背號,姓名,位置,體重,生日,經驗,大學
    players.append([year, team, no ,cols[0].text,cols[1].text,cols[3].text,cols[4].text,cols[6].text,cols[7].text])
  return players
```

```python
#生成time team url
def scraping(ttl):
  total_players= [['年份','球隊','背號','姓名','位置','體重','生日','經驗','大學']]
  for item in ttl:
    headers = {"user-agen" :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"}
    url = item[-1]
    team = item[0]
    year = item[1] 
    r = requests.get(url, headers=headers)
    print("抓取"+str(year)+"年的"+str(team+"資料"))
    if r.status_code == requests.codes.ok:
      soup = BeautifulSoup(r.text,"html.parser")
      team_players = players(soup,team,year)
      total_players = total_players+team_players
      time.sleep(5)
    else:
      print("http 請求錯誤")
  return total_players
```

```python
if __name__ == "__main__":
  url = "https://www.basketball-reference.com/teams/{0}/{1}.html"
  stop= "STOP"
  teams = []
  years = []
  team = str(input("請輸入要查找的隊伍（如不需要的請輸入stop)"))
  while team != stop:
    teams.append(team)
    team = str(input("請輸入要查找的隊伍（如不需要的請輸入stop)"))
  year = str(input("請輸入要查找的時間（如不需要的請輸入stop)"))  
  while year != stop:
    years.append(year)
    year = str(input("請輸入要查找的時間（如不需要的請輸入stop)"))
  ttl = gen_url(url,teams,years)
  total_players = scraping(ttl)
```
